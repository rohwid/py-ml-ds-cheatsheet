Start working with numpy operation

```python
import numpy as np

arr = np.arange(1,6)

arr
```

Arithmetic operation in numpy

```python
arr * arr
```

```python
arr - arr
```

```python
arr / arr
```

```python
1 / arr
```

```python
arr ** 2
```

```python
arr ** 3
```

# Average

```python
arr = np.arange(25)
arr = arr.reshape(5, 5)

arr
```

```python
np.average(arr)
```

```python
np.average(arr, axis=0)
```

```python
np.average(arr, axis=1)
```
