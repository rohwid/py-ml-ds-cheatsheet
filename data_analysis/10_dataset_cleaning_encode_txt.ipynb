```python
def convert_db_to_ner_dataset(inp, out):
    lines = open(inp).readlines()
    res_file = []

    for line in lines:
        res_file.append(line.encode().decode("unicode-escape"))

    open(out, 'w+').write(''.join(res_file))
```

```python
convert_db_to_ner_dataset('input/raw_encode.txt', 'output/raw_encoded.txt')
```
