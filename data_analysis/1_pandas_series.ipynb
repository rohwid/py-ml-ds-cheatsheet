# Series in Pandas

Start working with pandas and how initialize it

**Note:** So basically pandas is dictionary

```python
import numpy as np
import pandas as pd
```

## List and dictionary

Here's the example how to create series:
+ Assign python array to default pandas series
+ Assign python array to defined array as series
+ Assign numpy array to default pandas series
+ Assign numpy array to defined array as series

```python
labels = ['a','b','c']
my_list = [10, 20, 30]
d = {'a':10,'b':20,'c':30}
arr = np.array([10, 20, 30])
```

### Assign python array to default pandas series

Set my list array as list

```python
pd.Series(data=my_list)
```

### Assign python array to defined array as series (labels)

Set an array to be the other array label

```python
pd.Series(data=my_list, index=labels)
```

```python
pd.Series(my_list, labels)
```

### Assign numpy array to default pandas series

```python
pd.Series(arr)
```

### Assign numpy array to defined array as series (labels)

```python
pd.Series(arr, labels)
```

### Convert dictionary to series

```python
pd.Series(d)
```

## Data in Series

How to operate the data series (ex: arithmatic)

```python
pd.Series(data=labels)
```

```python
# Series param
pd.Series([sum, print, len])
```

```python
# How to use series and automatically assign data
ser1 = pd.Series([1,2,3,4], index = ['USA', 'Germany', 'USSR', 'Japan'])

ser1
```

```python
ser2 = pd.Series([1,2,5,4], index = ['USA', 'Germany', 'Italy', 'Japan'])

ser2
```

```python
ser1['USA']
```

```python
ser1 + ser2
```
