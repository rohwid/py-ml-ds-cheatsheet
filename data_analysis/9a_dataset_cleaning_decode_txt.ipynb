```python
def escape_line(res_sentence):
    return res_sentence.encode("unicode_escape").decode("utf-8")
    
def convert_ner_dataset_to_db(inp, out):
    lines = open(inp).readlines()
    res_file = []
    res_sentence = ''
    for line in lines:
        if line.strip() == '':
            res_sentence = res_sentence + line.strip()
            res_file.append(escape_line(res_sentence) + '\n')
            res_sentence = ''
        else: 
            res_sentence = res_sentence + line.strip() + '\n'
    res_file.append(escape_line(res_sentence))
    open(out, 'w+').write(''.join(res_file))
```

```python
convert_ner_dataset_to_db('input/raw_decode.txt', 'output/raw_decoded.txt')
```
