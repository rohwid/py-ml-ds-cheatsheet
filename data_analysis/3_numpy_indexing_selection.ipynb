### Array concatrnate

Join or merge two arrays into one array.

```python
a = [1, 2, 3, 4]
b = np.array([5, 6, 7, 8])
c = np.array([[1, 2], [3, 4]])
d = np.array([[5, 6]])

ab = np.concatenate((a, b), axis=None)

ab
```

Join a **number** to **an array**

```python
a = [1, 2, 3, 4]
b = np.array([5, 6, 7, 8])
c = np.array([[1, 2], [3, 4]])
d = np.array([[5, 6]])

number_b = np.concatenate((99, b), axis=None)

number_b
```

Join matrix **c** and **d** depends on **rows**

```python
a = [1, 2, 3, 4]
b = np.array([5, 6, 7, 8])
c = np.array([[1, 2], [3, 4]])
d = np.array([[5, 6]])

c
```

```python
d
```

```python
cd = np.concatenate((c, d), axis=0)

cd
```

Join matrix **c** and transposed matrix **d** depends on ***column***

```python
a = [1, 2, 3, 4]
b = np.array([5, 6, 7, 8])
c = np.array([[1, 2], [3, 4]])
d = np.array([[5, 6]])

cd = np.concatenate((c, d.T), axis=1)

cd
```

```python
a = [1, 2, 3, 4]
b = np.array([5, 6, 7, 8])
c = np.array([[1, 2], [3, 4]])
d = np.array([[5, 6]])

ab = np.concatenate((a, b), axis=None)

ab
```

Joint all list or arrays become an array, must be include `axis=None` syntax

```python
a = [1, 2, 3, 4]
b = np.array([5, 6, 7, 8])
c = np.array([[1, 2], [3, 4]])
d = np.array([[5, 6]])

# generalize all list or array as single row
abcd = np.concatenate((a, b, c, d), axis=None)

abcd
```

## Fancy Indexing

Little bit advancing the indexing and insert the values to it.

```python
arr2d = np.zeros((10,10))

arr2d
```

```python
arr_length = arr2d.shape[0]

arr_length
```

```python
for i in range(arr_length):
    arr2d[i] = i
    
arr2d
```

```python
arr2d[[2,4,6,8]]
```

```python
arr2d[[6,4,2,7]]
```

```python
arr = np.zeros((10,10))

arr
```

```python
len(arr)
```

```python
arr.shape
```

```python
arr = np.zeros((10,10))

for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
        if j == i:
            arr[i][j] = 7
            arr[(arr.shape[0] - 1) - j][j] = 7

arr
```

```python
arr = np.arange(100)
arr = arr.reshape(10, 10)

arr
```

Default looping in numpy is by row

```python
for col in arr:
    print(col)
    break
```

How to loop by column in numpy

```python
for col in arr.T:
    print(col)
    break
```
