## Check Array and NumpyArray

Here's some example of array checking for decission making:

```python
my_list = [1, 2, 3, 4, 5]
my_list_empty = []

arr = np.arange(15)
arr_empty = np.array([])
```

```python
if my_list == []:
    print("true")
else:
    print("false")
```

```python
if not my_list:
    print("true")
else:
    print("false")
```

```python
if my_list_empty == []:
    print("true")
else:
    print("false")
```

```python
if not my_list_empty:
    print("true")
else:
    print("false")
```

```python
if arr == []:
    print("true")
else:
    print("false")
```

```python
if arr_empty == []:
    print("true")
else:
    print("false")
```

```python
if arr.size == []:
    print("true")
else:
    print("false")
```

```python
if arr_empty.size == 0:
    print("true")
else:
    print("false")
```
